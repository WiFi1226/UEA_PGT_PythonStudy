# Pandas包

```python
import pandas as pd
pd.__version__
help(pd.read_excel) # 查看函数说明

# 数据选择
trains = pd.read_excel("https://github.com/tturocy/eco7026a/raw/main/2022-23/choochoo.xls")
df2 = trains.set_index('Year') # 字符串型需要加‘’，否则报错(如df2['Python'])。 如果加 ，inplace=True 代表同时更改源文件
# ? 为什么set_index()函数不需要没有截取时间戳？

# 数据类型
# ! Series类型-一维数组
# ! DataFrame类型-二维数组

trains.dtypes # 查看数据类型
type(df2['City']) # 变量即可

df2['City'] + df2['City'] # Series类型可直接相加,往下！

# 数据查看 行[] 列.loc[]
trains
trains['City']
trains[['City', 'Direction']]
trains.loc[591]






# 数据查找
# 注意数据类型str和float不可比
df2.describe() # 数据描述
df2.max() # 每行最大值，axis=0为列，axis=1为行，默认为0
df2.mean() # 每行均值
df2.value_counts() # 每列值计数
df2.sort_values(['','']) # 每列排序
df2.head() # 特定行

df2.groupby()[['','']] # 按列分组(Series类型)



df2.isnull() # 空值判断
df2.dropna() # 删除空值
df2.fillna() # 填充空值
df2.drop_duplicates() # 删除重复值
df2.drop() # 删除行/列

df2.mean() # 每列均值
df2.corr() # 相关系数
df2.cov() # 协方差
df2.apply() # 应用函数



df2.to_csv() # 输出csv文件
df2.to_excel() # 输出excel文件
df2.to_html() # 输出html文件
df2.to_json() # 输出json文件
df2.to_sql() # 输出sql文件
df2.to_clipboard() # 输出剪切板
df2.to_dict() # 输出字典
df2.to_feather() # 输出feather文件
df2.to_gbq() # 输出gbq文件

df2.aggregate(['','']) # 聚合函数，或(new_name_1='function1',new_name_2='function2')，或[('new_name_1','function1'),('new_name_2','function2')]
df2.transform() # 转换函数
df2.assign() # 赋值函数

df2.unqiue() # 唯一值函数
df2.nunique() # 唯一值计数函数

trains[''] == '' # 判断是否相等(输出类型为Series，实际为索引值列的列名null！！！！(无论是否比较，唯一标识为索引值列，输出为索引值列名null))——可用于筛选

# ! for Name in List: # for循环，name为变量(名字与List内容无关,取出的每一个值与List有关(i/name都可以)取出字符串用==比较)，list为列表
# ? ~符号的用法
```
